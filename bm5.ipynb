```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def SIR(u, t, beta, delta, epsilon):
    du = np.zeros(3)
    du[0] = epsilon*u[2]-beta*u[0]*u[1]
    du[1] = beta * u[0]*u[1] - delta * u[1]
    du[2] = delta * u[1] - epsilon*u[2]
    return du
```

```python
#CASO R0=5, 2 SEMANAS Y RECONTAGIO DE 0.5
R0 = 5
delta = 1.0 / 2  # inverso del tiempo que dura la enfermedad en semanas
beta = R0 * delta
epsilon = 1/10

t = np.linspace(0, 30, 2000)

u = odeint(SIR, [1, 0, 0], t, args=(beta, delta, epsilon))

plt.plot(t, u[:,0], label="S", color='green')
plt.plot(t, u[:,1], label="I", color='red')
plt.plot(t, u[:,2], label="R",color='orange')
plt.plot(t, np.sum(u, axis=1), label="T",color='brown')
plt.axhline(0, color="black", ls='--',lw=1,label= r'$I=0$')
plt.axhline(1, color="grey", ls='--',lw=1,label= r'$S=1$')
plt.legend()
plt.xlabel(r"t (Semanas)")
plt.ylabel(r"Población")
plt.title(r'Estado libre de enfermedad' )
plt.show()

u = odeint(SIR, [.99999, 0.00001, 0], t, args=(beta, delta, epsilon))
inf=epsilon*(R0-1)/(R0*(delta+epsilon))

plt.plot(t, u[:,0], label="S", color='green')
plt.plot(t, u[:,1], label="I", color='red')
plt.plot(t, u[:,2], label="R",color='orange')
plt.plot(t, np.sum(u, axis=1), label="T",color='brown')
plt.axhline(1/R0, color="lightgray", ls='--',lw=1, label= r'$S^*$')  #$S=1/R_0$
plt.axhline(inf, color="grey", ls='--',lw=1, label= r'$I^*$')   #$I=\epsilon(R_0-1)/R_0(\delta+\epsilon)$
plt.xlabel("t (Semanas)")
plt.ylabel("Población")
plt.title(r'Estado con enfermedad .001%' )
plt.legend()
plt.legend(loc='upper right')
plt.show()

u = odeint(SIR, [.99, 0.01, 0], t, args=(beta, delta, epsilon))

plt.plot(t, u[:,0], label="S", color='green')
plt.plot(t, u[:,1], label="I", color='red')
plt.plot(t, u[:,2], label="R",color='orange')
plt.plot(t, np.sum(u, axis=1), label="T",color='brown')
plt.axhline(1/R0, color="lightgray", ls='--',lw=1, label= r'$S^*$')  #$S=1/R_0$
plt.axhline(inf, color="grey", ls='--',lw=1, label= r'$I^*$')   #$I=\epsilon(R_0-1)/R_0(\delta+\epsilon)$
plt.xlabel("t (Semanas)")
plt.ylabel("Población")
plt.title(r'Estado con enfermedad 1%' )
plt.legend()
plt.legend(loc='upper right')
plt.show()

u = odeint(SIR, [.9, 0.1, 0], t, args=(beta, delta, epsilon))

plt.plot(t, u[:,0], label="S", color='green')
plt.plot(t, u[:,1], label="I", color='red')
plt.plot(t, u[:,2], label="R",color='orange')
plt.plot(t, np.sum(u, axis=1), label="T",color='brown')
plt.axhline(1/R0, color="lightgray", ls='--',lw=1, label= r'$S^*$')  #$S=1/R_0$
plt.axhline(inf, color="grey", ls='--',lw=1, label= r'$I^*$')   #$I=\epsilon(R_0-1)/R_0(\delta+\epsilon)$
plt.xlabel("t (Semanas)")
plt.ylabel("Población")
plt.title(r'Estado con enfermedad 10%' )
plt.legend()
plt.legend(loc='upper right')
plt.show()

u = odeint(SIR, [.75, 0.25, 0], t, args=(beta, delta, epsilon))

plt.plot(t, u[:,0], label="S", color='green')
plt.plot(t, u[:,1], label="I", color='red')
plt.plot(t, u[:,2], label="R",color='orange')
plt.plot(t, np.sum(u, axis=1), label="T",color='brown')
plt.axhline(1/R0, color="lightgray", ls='--',lw=1, label= r'$S^*$')  #$S=1/R_0$
plt.axhline(inf, color="grey", ls='--',lw=1, label= r'$I^*$')   #$I=\epsilon(R_0-1)/R_0(\delta+\epsilon)$
plt.xlabel("t (Semanas)")
plt.ylabel("Población")
plt.title(r'Estado con enfermedad 25%' )
plt.legend()
plt.legend(loc='upper right')
plt.show()

u = odeint(SIR, [0.5, 0.5, 0], t, args=(beta, delta, epsilon))

plt.plot(t, u[:,0], label="S", color='green')
plt.plot(t, u[:,1], label="I", color='red')
plt.plot(t, u[:,2], label="R",color='orange')
plt.plot(t, np.sum(u, axis=1), label="T",color='brown')
plt.axhline(1/R0, color="lightgray", ls='--',lw=1, label= r'$S^*$')  #$S=1/R_0$
plt.axhline(inf, color="grey", ls='--',lw=1, label= r'$I^*$')   #$I=\epsilon(R_0-1)/R_0(\delta+\epsilon)$
plt.xlabel("t (Semanas)")
plt.ylabel("Población")
plt.title(r'Estado con enfermedad 50%' )
plt.legend()
plt.legend(loc='upper right')
plt.show()

u = odeint(SIR, [0.2, 0.8, 0], t, args=(beta, delta, epsilon))

plt.plot(t, u[:,0], label="S", color='green')
plt.plot(t, u[:,1], label="I", color='red')
plt.plot(t, u[:,2], label="R",color='orange')
plt.plot(t, np.sum(u, axis=1), label="T",color='brown')
plt.axhline(1/R0, color="lightgray", ls='--',lw=1, label= r'$S^*$')  #$S=1/R_0$
plt.axhline(inf, color="grey", ls='--',lw=1, label= r'$I^*$')   #$I=\epsilon(R_0-1)/R_0(\delta+\epsilon)$
plt.xlabel("t (Semanas)")
plt.ylabel("Población")
plt.title(r'Estado con enfermedad 80%' )
plt.legend()
plt.legend(loc='upper right')
plt.show()
```

```python
def SIRV(u, t, beta, delta, epsilon, alfa):
    du = np.zeros(3)
    du[0] = epsilon*u[2]-beta*u[0]*u[1]-alfa*u[0]
    du[1] = beta * u[0]*u[1] - delta * u[1]
    du[2] = delta * u[1] - epsilon*u[2]+alfa*u[0]
    return du
```
