```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def Jin(t, ti, tf, h):
    if (t >= ti) and (t <= tf):
        aux = h
    else:
        aux = 0
    return aux

def two_comp_model(u, t, ti, tf, h, mu, gamma):
    du = np.zeros(2)
    du[0] = Jin(t, ti, tf, h) - mu * u[0] - gamma * (u[0] - u[1])
    du[1] = gamma * (u[0] - u[1])
    return du
```

```python
mu = np.log(2)/20.
t = np.linspace(0, 120, 2000)

gamma = 0
ti = 10
tf = 30
h = 5
u = odeint(two_comp_model, [0, 0], t, args=(ti, tf, h, mu, gamma))
plt.plot(t, u[:,0])


gamma = .1*mu
ti = 10
tf = 30
h = 5
u = odeint(two_comp_model, [0, 0], t, args=(ti, tf, h, mu, gamma))
plt.plot(t, u[:,0])

gamma = mu
ti = 10
tf = 30
h = 5
u = odeint(two_comp_model, [0, 0], t, args=(ti, tf, h, mu, gamma))
plt.plot(t, u[:,0])

gamma = 1000*mu
ti = 10
tf = 30
h = 5
u = odeint(two_comp_model, [0, 0], t, args=(ti, tf, h, mu, gamma))
plt.plot(t, u[:,0])
```

# PERIODICO

```python
def Jin_per(t, ti, T, dt, h):
    n = (t-ti)//T
    if (t - ti - n*T) <= dt:
        aux = h
    else:
        aux = 0
    return aux

def two_comp_model_per(u, t, ti, T, dt, h, mu, gamma):
    du = np.zeros(2)
    du[0] = Jin_per(t, ti, T, dt, h) - mu * u[0] - gamma * (u[0] - u[1])
    du[1] = gamma * (u[0] - u[1])
    return du
```

```python
#Concentracion en la sangre 
ti = 0
T = 12  #horas de dosis
dt = 4 #tiempo 
h = 25 #dosis
tmed=22 #vida media
mu = np.log(2)/tmed 

t = np.linspace(0,360 , 12000)

gamma = 0
u = odeint(two_comp_model_per, [0, 0], t, args=(ti, T, dt, h, mu, gamma))
plt.plot(t,u[:,0],label='$\gamma=0$')

gamma = 0.1*mu
u = odeint(two_comp_model_per, [0, 0], t, args=(ti, T, dt, h, mu, gamma))
plt.plot(t,u[:,0],label='$\gamma=0.1\mu$')

gamma = mu
u = odeint(two_comp_model_per, [0, 0], t, args=(ti, T, dt, h, mu, gamma))
plt.plot(t,u[:,0],label='$\gamma=\mu$')

gamma = 10*mu
u = odeint(two_comp_model_per, [0, 0], t, args=(ti, T, dt, h, mu, gamma))
plt.plot(t,u[:,0],label='$\gamma=10\mu$')

plt.ylabel('Concentración')
plt.xlabel("Tiempo (horas)")
plt.title(r'Concentración en Plasma, dosis 25 mg' )

plt.legend()
plt.show()
```
