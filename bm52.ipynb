```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def SIR(u, t, beta, delta, epsilon):
    du = np.zeros(3)
    du[0] = epsilon*u[2]-beta*u[0]*u[1]
    du[1] = beta * u[0]*u[1] - delta * u[1]
    du[2] = delta * u[1] - epsilon*u[2]
    return du
```

```python
#CASO R0=5, 2 SEMANAS Y RECONTAGIO DE 0.5
R0 = 5
delta = 1.0 / 2  # inverso del tiempo que dura la enfermedad en semanas
beta = R0 * delta
epsilon = 1/10

t = np.linspace(0, 30, 2000)

p0=[1, 0, 0]
u = odeint(SIR, p0, t, args=(beta, delta, epsilon))
plt.figure(figsize=(10, 6))
plt.plot(u[:, 0], u[:, 1],color="skyblue")


u = odeint(SIR, [.99999, 0.00001, 0], t, args=(beta, delta, epsilon))
plt.plot(u[:, 0], u[:, 1],color="skyblue")


u = odeint(SIR, [.99, 0.01, 0], t, args=(beta, delta, epsilon))
plt.plot(u[:, 0], u[:, 1],color="skyblue")


u = odeint(SIR, [.9, 0.1, 0], t, args=(beta, delta, epsilon))
plt.plot(u[:, 0], u[:, 1],color="skyblue")


u = odeint(SIR, [.75, 0.25, 0], t, args=(beta, delta, epsilon))
plt.plot(u[:, 0], u[:, 1],color="skyblue")


u = odeint(SIR, [0.5, 0.5, 0], t, args=(beta, delta, epsilon))
plt.plot(u[:, 0], u[:, 1],color="skyblue")


u = odeint(SIR, [0.2, 0.8, 0], t, args=(beta, delta, epsilon))
plt.plot(u[:, 0], u[:, 1],color="skyblue")


plt.plot([1,1/R0],[0,epsilon*(R0-1)/(R0*(delta+epsilon))],'o',color="darkred")
plt.ylabel(r"Poblacion i ")
plt.xlabel(r"Poblacion s ")
plt.title(r'Espacio Fase R0>1' )
plt.show()

```

```python
def SIRV(u, t, beta, delta, epsilon, alfa):
    du = np.zeros(3)
    du[0] = epsilon*u[2]-beta*u[0]*u[1]-alfa*u[0]
    du[1] = beta * u[0]*u[1] - delta * u[1]
    du[2] = delta * u[1] - epsilon*u[2]+alfa*u[0]
    return du
```

```python
#CASO R0=5, 2 SEMANAS Y RECONTAGIO DE 1/10 vacuna .00001
R0 = 5
delta = 1.0 / 2  # inverso del tiempo que dura la enfermedad en semanas
beta = R0 * delta
epsilon = 1/10
alfa=0.01

t = np.linspace(0, 30, 2000)

u = odeint(SIRV, [1, 0, 0], t, args=(beta, delta, epsilon,alfa))
plt.plot(u[:, 0], u[:, 1],color="pink")

u = odeint(SIRV, [.999, 0.0005, 0.0005], t, args=(beta, delta, epsilon,alfa))
plt.plot(u[:, 0], u[:, 1],color="pink")


u = odeint(SIRV, [.99, 0.007, 0.003], t, args=(beta, delta, epsilon, alfa))
plt.plot(u[:, 0], u[:, 1],color="pink")


u = odeint(SIRV, [.9, 0.06, 0.04], t, args=(beta, delta, epsilon,alfa))
plt.plot(u[:, 0], u[:, 1],color="pink")


u = odeint(SIRV, [.75, 0.15, 0.10], t, args=(beta, delta, epsilon,alfa))
plt.plot(u[:, 0], u[:, 1],color="pink")


u = odeint(SIRV, [0.5, 0.5, 0], t, args=(beta, delta, epsilon,alfa))
plt.plot(u[:, 0], u[:, 1],color="pink")


u = odeint(SIRV, [0.2, 0.6, 0.2], t, args=(beta, delta, epsilon,alfa))
plt.plot(u[:, 0], u[:, 1],color="pink")

plt.plot([epsilon/(epsilon+alfa),1/R0],[0,(epsilon*R0-epsilon-alfa)/(R0*(delta+epsilon))],'o',color="darkred")
plt.ylabel(r"Poblacion i ")
plt.xlabel(r"Poblacion s ")
plt.title(r'Espacio Fase R0>1' )
plt.show()
```
